```python
# NOTEBOOK: 04_Orchestration
# OBJETIVO: Coordenar a execução de todos os notebooks na ordem correta

# ===========================================================================
# POR QUE ORQUESTRAR?
# ===========================================================================
# 1. GARANTIR ORDEM: Bronze → Silver → Gold (dependências)
# 2. FACILITAR RE-EXECUÇÃO: Um comando para rodar tudo
# 3. PREPARAR PARA PRODUÇÃO: Workflows agendados
# ===========================================================================

# Lista de notebooks na ORDEM CORRETA
notebooks = [
    {"path": "01_ETL_Medallion_Architecture", "timeout": 300, "nome": "Bronze Layer"},
    {"path": "03_Data_Quality_Checks", "timeout": 120, "nome": "Data Quality Bronze"},
    {"path": "02_Silver_Transformation", "timeout": 300, "nome": "Silver Layer"},
    {"path": "03_Data_Quality_Checks", "timeout": 120, "nome": "Data Quality Silver"},
    {"path": "03_Gold_Dimensional_Model", "timeout": 300, "nome": "Gold Layer"},
    {"path": "03_Data_Quality_Checks", "timeout": 120, "nome": "Data Quality Gold"}
]

print(" INICIANDO PIPELINE COMPLETO...")

for notebook in notebooks:
    try:
        print(f"▶ Executando: {notebook['nome']}")
        
        # dbutils.notebook.run() é a função do Databricks para executar notebooks
        result = dbutils.notebook.run(notebook["path"], notebook["timeout"])
        
        print(f" {notebook['nome']} - SUCESSO")
        print(f"   Resultado: {result}")
        
    except Exception as e:
        print(f" {notebook['nome']} - ERRO: {str(e)}")
        
        # EM PRODUÇÃO: Aqui você poderia:
        # - Enviar email de alerta
        # - Registrar em sistema de monitoring
        # - Tomar decisão de continuar ou parar
        
        raise e  # Para o pipeline em caso de erro

print(" PIPELINE CONCLUÍDO COM SUCESSO!")
print(" Tabelas Gold prontas para consumo no Power BI")
```
