```python
%sql
-- NOTEBOOK: 00_Setup_Logging 
-- OBJETIVO: Criar as tabelas de log

USE CATALOG workspace;
USE pl_delivery_analysis;

-- ===========================================================================
-- CRIAR TABELAS COM ESTRUTURA CORRETA
-- ===========================================================================

-- Remover tabelas existentes (se houver) para recriar com estrutura correta
DROP TABLE IF EXISTS log_processamento;
DROP TABLE IF EXISTS metricas_qualidade;

-- Tabela de logs de processamento - COM COLUNA DETALHES
CREATE TABLE log_processamento (
    id BIGINT GENERATED ALWAYS AS IDENTITY,
    processo STRING,
    etapa STRING,
    registros_afetados INT,
    status STRING,
    timestamp TIMESTAMP,
    duracao_segundos INT,
    detalhes STRING
);

-- Tabela para m√©tricas de qualidade
CREATE TABLE metricas_qualidade (
    id BIGINT GENERATED ALWAYS AS IDENTITY,
    metrica STRING,
    valor DECIMAL(15,2),
    data_calculo TIMESTAMP
);

-- Verificar a estrutura da tabela log_processamento
DESCRIBE log_processamento;

SELECT ' Tabelas de log recriadas com estrutura correta' as status;
```
